```python
# XAI Methods for Transformer-based Models in NLP
# This notebook demonstrates three XAI techniques:
# 1. Attention Visualization
# 2. Gradient-based Saliency Mapping
# 3. LIME Explanations

import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from transformers import AdamW, get_linear_schedule_with_warmup
from sklearn.metrics import accuracy_score, f1_score
from torch.utils.data import Dataset, DataLoader
import lime
import lime.lime_text
import captum
from captum.attr import LayerIntegratedGradients, visualization

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# 1. Data Preparation
# For fast training, we'll use a subset of the IMDB dataset
# This is easily available through Kaggle datasets

def load_data():
    # You can use this with Kaggle's IMDB dataset
    # Just add the dataset to your notebook: kaggle datasets download -d lakshmi25npathi/imdb-dataset-of-50k-movie-reviews
    try:
        df = pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
        # Take a small subset for quick training
        df_train = df.sample(2000, random_state=42)
        df_val = df[~df.index.isin(df_train.index)].sample(500, random_state=42)
        
        # Convert sentiment to numeric labels
        df_train['label'] = df_train['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
        df_val['label'] = df_val['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
        
        return df_train, df_val
    except:
        # Fallback to a toy dataset for demonstration
        from sklearn.datasets import fetch_20newsgroups
        categories = ['alt.atheism', 'soc.religion.christian']
        newsgroups_train = fetch_20newsgroups(subset='train', categories=categories)
        newsgroups_test = fetch_20newsgroups(subset='test', categories=categories)
        
        df_train = pd.DataFrame({
            'review': newsgroups_train.data[:1000],
            'label': newsgroups_train.target[:1000]
        })
        df_val = pd.DataFrame({
            'review': newsgroups_test.data[:200],
            'label': newsgroups_test.target[:200]
        })
        return df_train, df_val

# Dataset class
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        
        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }

# 2. Model Training Function
def train_model(model, train_dataloader, val_dataloader, epochs=3):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    model = model.to(device)
    
    optimizer = AdamW(model.parameters(), lr=2e-5)
    total_steps = len(train_dataloader) * epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=0,
        num_training_steps=total_steps
    )
    
    loss_fn = torch.nn.CrossEntropyLoss()
    
    best_accuracy = 0
    
    for epoch in range(epochs):
        print(f'Epoch {epoch + 1}/{epochs}')
        
        model.train()
        train_losses = []
        
        for batch in tqdm(train_dataloader):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)
            
            model.zero_grad()
            
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask
            )
            
            loss = loss_fn(outputs.logits, labels)
            train_losses.append(loss.item())
            
            loss.backward()
            optimizer.step()
            scheduler.step()
        
        print(f'Train loss: {np.mean(train_losses)}')
        
        # Validation
        model.eval()
        val_losses = []
        predictions = []
        true_labels = []
        
        with torch.no_grad():
            for batch in tqdm(val_dataloader):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['label'].to(device)
                
                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask
                )
                
                loss = loss_fn(outputs.logits, labels)
                val_losses.append(loss.item())
                
                _, preds = torch.max(outputs.logits, dim=1)
                predictions.extend(preds.cpu().tolist())
                true_labels.extend(labels.cpu().tolist())
        
        val_loss = np.mean(val_losses)
        accuracy = accuracy_score(true_labels, predictions)
        f1 = f1_score(true_labels, predictions, average='weighted')
        
        print(f'Validation loss: {val_loss}')
        print(f'Accuracy: {accuracy}')
        print(f'F1 Score: {f1}')
            
    return model

# 3. XAI Methods Implementation

# 3.1 Attention Visualization
def visualize_attention(model, tokenizer, text, label_names=['Negative', 'Positive']):
    """Visualize attention weights for a given text"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()
    
    encoding = tokenizer(
        text,
        return_tensors='pt',
        padding=True,
        truncation=True,
        max_length=128
    )
    
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)
    
    with torch.no_grad():
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            output_attentions=True
        )
    
    # Get prediction
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=1).item()
    
    # Get attention weights from the last layer
    # We're using the CLS token's attention to all other tokens
    attentions = outputs.attentions[-1].detach().cpu().numpy()
    
    # Extract attention from CLS token ([0, 0] for batch 0, first token)
    cls_attention = attentions[0, 0, 0, :]
    
    # Get tokens
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
    
    # Plot attention weights
    plt.figure(figsize=(10, 6))
    sns.barplot(x=cls_attention[:len(tokens)], y=tokens)
    plt.title(f'Attention Weights for "{text[:50]}..." (Predicted: {label_names[predicted_class]})')
    plt.tight_layout()
    plt.show()
    
    return {
        'prediction': label_names[predicted_class],
        'tokens': tokens,
        'attention_weights': cls_attention[:len(tokens)]
    }

# 3.2 Gradient-based Saliency Mapping using Captum
def token_erasure_attribution(model, tokenizer, text, label_names=['Negative', 'Positive']):
    """
    Compute token importance by erasing each token and measuring impact on prediction
    """
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()
    
    # Tokenize the input text
    encoding = tokenizer(
        text,
        return_tensors='pt',
        padding=True,
        truncation=True,
        max_length=128
    )
    
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)
    
    # Get the original prediction and confidence
    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        
    original_probs = torch.nn.functional.softmax(outputs.logits, dim=1)
    predicted_class = torch.argmax(original_probs, dim=1).item()
    original_confidence = original_probs[0, predicted_class].item()
    
    # Find tokens to analyze (skip special tokens)
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
    token_ids = input_ids[0].cpu().numpy()
    
    # Calculate importance by erasing each token
    token_importance = []
    
    for i in range(len(tokens)):
        # Skip special tokens like [CLS], [SEP], [PAD]
        if tokens[i] in tokenizer.special_tokens_map.values():
            token_importance.append(0)
            continue
            
        # Create a modified input by replacing the token with [MASK] or [UNK]
        modified_input_ids = input_ids.clone()
        modified_input_ids[0, i] = tokenizer.mask_token_id  # or tokenizer.unk_token_id
        
        # Get new prediction
        with torch.no_grad():
            modified_outputs = model(
                input_ids=modified_input_ids, 
                attention_mask=attention_mask
            )
            
        modified_probs = torch.nn.functional.softmax(modified_outputs.logits, dim=1)
        modified_confidence = modified_probs[0, predicted_class].item()
        
        # Importance = drop in confidence when token is removed
        importance = original_confidence - modified_confidence
        token_importance.append(importance)
    
    # Convert to numpy for visualization
    token_importance = np.array(token_importance)
    
    # Normalize importance scores
    if token_importance.max() > 0:
        token_importance = token_importance / token_importance.max()
    
    # Plot attributions
    plt.figure(figsize=(10, 6))
    sns.barplot(x=token_importance, y=tokens)
    plt.title(f'Token Importance for "{text[:50]}..." (Predicted: {label_names[predicted_class]})')
    plt.tight_layout()
    plt.show()
    
    return {
        'prediction': label_names[predicted_class],
        'tokens': tokens,
        'attributions': token_importance
    }

# 3.3 LIME Explanation
def lime_explanation(model, tokenizer, text, label_names=['Negative', 'Positive']):
    """Generate LIME explanations for model predictions"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()
    
    # Define prediction function for LIME
    def predict_proba(texts):
        encoded_texts = tokenizer(
            texts,
            padding=True,
            truncation=True,
            max_length=128,
            return_tensors='pt'
        )
        
        input_ids = encoded_texts['input_ids'].to(device)
        attention_mask = encoded_texts['attention_mask'].to(device)
        
        with torch.no_grad():
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask
            )
        
        probs = torch.nn.functional.softmax(outputs.logits, dim=1)
        return probs.cpu().numpy()
    
    # Initialize LIME explainer
    explainer = lime.lime_text.LimeTextExplainer(class_names=label_names)
    
    # Generate explanation
    explanation = explainer.explain_instance(
        text,
        predict_proba,
        num_features=10,
        num_samples=100
    )
    
    # Get prediction
    probs = predict_proba([text])[0]
    predicted_class = np.argmax(probs)
    
    # Display explanation
    print(f'Text: "{text[:100]}..."')
    print(f'Predicted class: {label_names[predicted_class]} ({probs[predicted_class]:.4f})')
    
    # Plot LIME explanation
    plt.figure(figsize=(10, 6))
    explanation.as_pyplot_figure()
    plt.tight_layout()
    plt.show()
    
    return {
        'prediction': label_names[predicted_class],
        'probability': probs[predicted_class],
        'explanation': explanation
    }

# 4. Main Execution
def main():
    print("Loading data...")
    df_train, df_val = load_data()
    
    print("Loading pre-trained model and tokenizer...")
    model_name = 'distilbert-base-uncased'  # Smaller than BERT, faster to train
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=2,
        output_attentions=True
    )
    
    print("Preparing datasets...")
    train_dataset = TextDataset(
        texts=df_train['review'].values,
        labels=df_train['label'].values,
        tokenizer=tokenizer
    )
    
    val_dataset = TextDataset(
        texts=df_val['review'].values,
        labels=df_val['label'].values,
        tokenizer=tokenizer
    )
    
    train_dataloader = DataLoader(
        train_dataset,
        batch_size=16,
        shuffle=True
    )
    
    val_dataloader = DataLoader(
        val_dataset,
        batch_size=16,
        shuffle=False
    )
    
    print("Training model...")
    # Set epochs=1 for fast execution, increase for better performance
    model = train_model(model, train_dataloader, val_dataloader, epochs=10)

    # Save the model and tokenizer using save_pretrained
    output_dir = "./distilbert_sentiment_model"
    print(f"Saving model and tokenizer to {output_dir}...")
    model.save_pretrained(output_dir)
    tokenizer.save_pretrained(output_dir)
    print(f"Model and tokenizer saved to {output_dir}")
    
    # Demonstrate XAI methods on a sample text
    sample_text = "This movie was absolutely fantastic. The acting was superb and the plot was engaging from start to finish."
    
    print("\n===== Attention Visualization =====")
    attention_results = visualize_attention(model, tokenizer, sample_text)
    
    print("\n===== Gradient-based Attribution =====")
    gradient_results = token_erasure_attribution(model, tokenizer, sample_text)
    
    print("\n===== LIME Explanation =====")
    lime_results = lime_explanation(model, tokenizer, sample_text)
    
    # Compare results from different XAI methods
    print("\n===== XAI Methods Comparison =====")
    print("Predicted class:", attention_results['prediction'])
    
    # Print top 5 tokens by each method
    print("\nTop tokens by attention:")
    attn_indices = np.argsort(attention_results['attention_weights'])[-5:][::-1]
    for idx in attn_indices:
        print(f"- {attention_results['tokens'][idx]}: {attention_results['attention_weights'][idx]:.4f}")
    
    print("\nTop tokens by gradient attribution:")
    grad_indices = np.argsort(gradient_results['attributions'])[-5:][::-1]
    for idx in grad_indices:
        print(f"- {gradient_results['tokens'][idx]}: {gradient_results['attributions'][idx]:.4f}")
    
    print("\nTop features by LIME:")
    lime_features = lime_results['explanation'].as_list()
    for feature, weight in lime_features[:5]:
        print(f"- {feature}: {weight:.4f}")

if __name__ == "__main__":
    main()
```
